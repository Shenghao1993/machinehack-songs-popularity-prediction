## 1.1 Convert "Likes" and "Popularity" to numeric data type

```python
def conv_numeric_feature(feature):
    feature = feature.replace(',', '')
    vals = [int(val) for val in re.findall('\d+', feature)]
    if len(vals) == 1:
        return vals[0] * 1000 if 'K' in feature else vals[0] 
    elif len(vals) == 2:
        if '.' in feature:
            value = vals[0] + 0.1 * vals[1]
            if 'K' in feature:
                value = value * 1000
            return value
        return None
    else:
        return None
```

```python
train_df['Likes'] = train_df['Likes'].apply(lambda likes: conv_numeric_feature(likes))
train_df['Popularity'] = train_df['Popularity'].apply(lambda popularity: conv_numeric_feature(popularity))
test_df['Likes'] = test_df['Likes'].apply(lambda likes: conv_numeric_feature(likes))
test_df['Popularity'] = test_df['Popularity'].apply(lambda popularity: conv_numeric_feature(popularity))
```

# III. Feature Engineering
First, remove outliers with high ratio of views over likes.

```python
train_df_outliers_removed = train_df[train_df['BC_Views'] / train_df['BC_Likes'] < 10]
print(train_df_outliers_removed.shape)
train_df_outliers_removed.head()
```

## 3.1 Create Categories for Genres
Genres will be split into 4 groups by their frequency and distribution of view counts. One purpose of grouping the genres is to overcome the class imbalance. The resulting 4 genre groups are as follows.
<ul>
    <li>indie, hiphoprap, classical, all-music, reggaeton, dubstep, rbsoul, drumbass, disco</li>
    <li>danceedm</li>
    <li>electronic</li>
    <li>alternativerock, country, metal, deephouse, trap, latin, ambient, pop, rock, folksingersongwriter</li>
</ul>

```python
group0 = ['indie', 'hiphoprap', 'classical', 'all-music', 'reggaeton', 'dubstep', 'rbsoul', 'drumbass', 'disco']
group3 = ['alternativerock', 'country', 'metal', 'deephouse', 'trap',
          'latin', 'ambient', 'pop', 'rock', 'folksingersongwriter']
train_df_outliers_removed.loc[train_df_outliers_removed['Genre'].isin(group0), 'Categorical_Genre'] = 0
train_df_outliers_removed.loc[train_df_outliers_removed['Genre'] == 'danceedm', 'Categorical_Genre'] = 1
train_df_outliers_removed.loc[train_df_outliers_removed['Genre'] == 'electronic', 'Categorical_Genre'] = 2
train_df_outliers_removed.loc[train_df_outliers_removed['Genre'].isin(group3), 'Categorical_Genre'] = 3
```

## 3.2 Create Categories for Release Year & Month

```python
train_df_outliers_removed.loc[train_df_outliers_removed['Release_Year'] < 2016, 'Categorical_Year'] = 0
train_df_outliers_removed.loc[train_df_outliers_removed['Release_Year'].isin([2016, 2017]), 'Categorical_Year'] = 1
train_df_outliers_removed.loc[train_df_outliers_removed['Release_Year'] > 2017, 'Categorical_Year'] = 2
train_df_outliers_removed.loc[train_df_outliers_removed['Release_Month'].isin([12, 1, 2]), 'Categorical_Month'] = 0
train_df_outliers_removed.loc[train_df_outliers_removed['Release_Month'].isin([3, 4, 5]), 'Categorical_Month'] = 1
train_df_outliers_removed.loc[train_df_outliers_removed['Release_Month'].isin([6, 7, 8]), 'Categorical_Month'] = 2
train_df_outliers_removed.loc[train_df_outliers_removed['Release_Month'].isin([9, 10, 11]), 'Categorical_Month'] = 3

cols_to_drop = ['Unique_ID', 'Name', 'Genre', 'Country', 'Song_Name', 'Timestamp',
                'Views', 'Comments', 'Likes', 'Popularity', 'Followers', 'Release_Year', 'Release_Month']
train_df_outliers_removed.drop(columns=cols_to_drop, inplace=True)
train_df_outliers_removed['Categorical_Genre'] = train_df_outliers_removed['Categorical_Genre'].astype(int)
train_df_outliers_removed['Categorical_Year'] = train_df_outliers_removed['Categorical_Year'].astype(int)
train_df_outliers_removed['Categorical_Month'] = train_df_outliers_removed['Categorical_Month'].astype(int)
train_df_outliers_removed.head()
```

```python
encoded_train_df = pd.get_dummies(train_df_outliers_removed,
                                  columns=['Categorical_Genre', 'Categorical_Year', 'Categorical_Month'])
print(encoded_train_df.shape)
encoded_train_df.head()
```

Process training and test data and export to files.

```python
preprocessor = Preprocessor("data/Data_Train.csv", "data/Data_Test.csv")
processed_train_df, processed_test_df, boxcox_lambdas = preprocessor.run()
print("Dimension of training data: ", processed_train_df.shape)
print("Dimension of test data: ", processed_test_df.shape)

processed_train_df.to_csv("output/Processed_Train.csv", index=False)
processed_test_df.to_csv("output/Processed_Test.csv", index=False)
with open('output/boxcox_lambdas.json', 'w') as f:
    json.dump(boxcox_lambdas, f, indent=4)
```
